```python
# Removing the unwanted text from the ratings columns and also changing its data type to int
df["Rating"] = df["Rating"].str.replace(" Ratings", "").str.replace(",", "").astype("int")
```

```python
# Renmaed the columns from Reviwes to Overall Reviews
df = df.rename(columns={"Reviews":"Overall_Reviews"})
```

```python
df
```

```python
# Removing the unwanted text from the overall reviews columns and also changing its data type to int
df["Overall_Reviews"]=df["Overall_Reviews"].str.replace("Reviews","").str.replace(",", "").astype("int")
```

```python
df
```

```python
# After that checking the data type which we have changed 
df.info()
```

```python
# Removing the unwanted text from tha Actual Price Columns 
df["Actual price"] = pd.to_numeric(df["Actual price"].str.replace("₹", "").str.replace(",", ""), errors='coerce')
```

```python
# Filling the Missing values as 0.0
df["Actual price"] =df["Actual price"] .fillna(0.0)
```

```python
# Converting the data type to int
df["Actual price"]  = df["Actual price"].astype("int")
```

```python
# Removing the unwanted text from tha Discount Price Columns 
df["Discount price"] = pd.to_numeric(df["Discount price"].str.replace("₹", "").str.replace(",", ""), errors='coerce')
```

```python
# Filling the Missing values as 0.0
df["Discount price"] = df["Discount price"].fillna(0.0)
```

```python
# Converting the data type to int
df["Discount price"] = df["Discount price"].astype("int")
```

```python
# After that checking the data type which we have changed 
df.info()
```
